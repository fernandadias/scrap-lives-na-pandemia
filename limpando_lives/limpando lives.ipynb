# Script para a limpar o dataset das lives no YouTube
Dataset criado a partir do script em: https://github.com/fernandadias/scrap-lives-na-pandemia

```python
import pandas as pd

# importar o dataset
lives = pd.read_csv('pegar_lives.csv')
lives.head(4)
```

## Separando as lives
Percebemos que as lives vieram com um texto junto a data dizendo 'Streamed live'.
Usamos isso para fazer a primeira limpeza dos resultados, separando os streams dos outros videos.

```python
# string que define as lives
lives_string ='Streamed live'
 
# sinalizando as lives no df
lives["live"]= lives["data_publicacao"].str.contains(lives_string)

# selecionando só as lives
lives = lives[lives["live"] == True]

lives.reset_index(drop=True, inplace=True)
lives.head(4)
```

## Tratando os demais dados
Limpando as quebras de linha do scrap e normalizando os demais textos (datas e números)

```python
# limpando os espaços
lives = lives.replace('\n',' ', regex=True)
```

```python
# tratando as visualizações
lives['visualizacoes'].describe()
```

```python
lives['visualizacoes'] = lives['visualizacoes'].str.replace(r' views', '')
lives['visualizacoes'] = lives['visualizacoes'].str.replace(r',', '')
lives.head(4)
```

```python
lives.sort_values(by=['visualizacoes'])
```

```python
from datetime import datetime

lives['data_publicacao'] = lives['data_publicacao'].str.replace(r'Premiered ', '')
lives['data_publicacao'] = lives['data_publicacao'].str.replace(r'Streamed live on ', '')
lives['data_publicacao'] = lives['data_publicacao'].str.replace(r'Streamed live ', '')

lives['data_publicacao'] = pd.to_datetime(lives['data_publicacao'], format='%b %d, %Y',errors='coerce')
lives
```

```python
lives.sort_values(by=['data_publicacao'], ascending=False)
```

```python
# limpando inscritos
lives['inscritos'] = lives['inscritos'].str.replace(r'subscribers', '')

lives
```

```python
# convertendo as quantidades

lives['likes'] = lives['likes'].str.replace('K','000')
lives['likes'] = lives['likes'].str.replace('M','000000')
lives['likes'] = lives['likes'].str.replace('.','', regex=True)

lives['deslikes'] = lives['deslikes'].str.replace('K','000')
lives['deslikes'] = lives['deslikes'].str.replace('M','000000')
lives['deslikes'] = lives['deslikes'].str.replace('.','', regex=True)

lives['inscritos'] = lives['inscritos'].str.replace('K','000')
lives['inscritos'] = lives['inscritos'].str.replace('M','000000')
lives['inscritos'] = lives['inscritos'].str.replace('.','', regex=True)

lives
```

```python
lives_export = lives.drop(['Page_Title', 'comentarios', 'Meta_description', 'Meta_keywords' ], axis=1)
```

```python
lives_export.to_csv('lives_export.csv', index=False) 
```
